```python
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from sklearn.metrics import accuracy_score
import nltk
```

# Load the preprocessed headlines


```python

processed_headlines_df = pd.read_csv('../database/processed_headlines.csv')
processed_headlines = processed_headlines_df['Processed_Headline'].tolist()
```

# Initialize the VADER sentiment analyzer

```python
nltk.download('vader_lexicon')
vader_analyzer = SentimentIntensityAnalyzer()
```

```python
def get_vader_sentiment(text):
    scores = vader_analyzer.polarity_scores(text)
    compound_score = scores['compound']
    if compound_score >= 0.05:
        return 'positive'
    elif compound_score <= -0.05:
        return 'negative'
    else:
        return 'neutral'
```

```python
vader_sentiments = [get_vader_sentiment(headline) for headline in processed_headlines]
```

```python
processed_headlines_df['VADER_Sentiment'] = vader_sentiments
processed_headlines_df.to_csv('vader.csv', index=False)
```
